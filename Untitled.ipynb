# STACK AND UNSTACK

```python
mw = df1[(df1['Edition']==2008)&((df1['Event']=='100m')|(df1['Event']=='200m'))]
```

```python
g_mw = mw.groupby(by=['NOC','Gender','Discipline','Event']).size()
g_mw
```

```python
df = g_mw.unstack([ 'Discipline','Event'])
df
```

```python
df.stack('Event')
```

```python
x = df1[(df1.NOC == 'USA')].groupby(by=['Edition', 'Athlete', 'Discipline'], group_keys=False).agg(
total=('Medal','count')
)
x
```
